```python
def count_city_wise(products):
    #products is a call to search_for_products
    experiences = products["data"]
    city_wise_count = {} #{city_name: {count, RxN, total_reviews, price_total}}
    for data in experiences: #average price, priceFormatted - price, weighted rating - reviewCount, rating, primaryDestinationName
        if data["primaryDestinationName"] in city_wise_count.keys():
            city_wise_count[data["primaryDestinationName"]]["count"] += 1
            city_wise_count[data["primaryDestinationName"]]["Rating"] += data["rating"]
            city_wise_count[data["primaryDestinationName"]]["total_reviews"] += data["reviewCount"]
            city_wise_count[data["primaryDestinationName"]]["price_total"] += data["price"]
            city_wise_count[data["primaryDestinationName"]]["primaryDestinationId"] = data["primaryDestinationId"]
        else:
            city_wise_count[data["primaryDestinationName"]] = {"count":1,
                                                              "Rating": data["rating"],
                                                              "total_reviews": data["reviewCount"],
                                                              "price_total": data["price"],
                                                              "primaryDestinationId":  data["primaryDestinationId"]}
    return city_wise_count
```

## Flow of the run
1. For each country, count all the experiences of each city. (count_city_wise)
2. The city which has most number of experience is the hottest.
3. Sort the cities based on no. of experiences.
4. Now for each city, take the top 10 experience (all details of those experiences) based on average rating (top_20_products_to_gsheet)
    * Now here the thing comes of matching the experiences with Google maps or other sources and taking the best 10 experience out.
5. These top 10 experience of each city for a country is found.
6. The for each experience for a city product info (all info about experience) is saved in mongoDB atlas. (product_info)
#### Till here the work is in this notebook. After this work is in [MongoDB Atlas.ipynb](http://localhost:8888/notebooks/MongoDB%20Atlas.ipynb)
6. After all databases are saved, then each db is retrieved and for each city, the following points are retrieved. [like in this sheet](https://docs.google.com/spreadsheets/d/1ITNQkwRAr7FNPybq9gD4MgfbzfMlC6S3gMjRV3E2vZ8/edit#gid=0)
    * No. Of counts below 12 hours
    * Average cost per person per experience (INR)
    * Average duration of experience (In hours)
    * Number of persons
    * Number of nights
    * Multiplicative Factor
    * Starting Hotel price
    * Average Flight price
7. And then finally the total price after computation is returned.
8. Some other points to add are Best time to visit.

```python
def format_duration(duration):
    try:
        #type of '6 hours', '1 hour', '6 hours 30 minutes', '1 hour 30 minutes'
        hours = int(duration.split(' hour')[0])
        return(timedelta(hours = hours))        
    except:
        try:
            #type of '4 to 5 hours'
            hours_1 = int(duration.split(' hours')[0].split(' to ')[0])
            hours_2 = int(duration.split(' hours')[0].split(' to ')[1]) 
            return(timedelta(hours = hours_2))
        except:
            try:
                #type '3 days'
                days = int(duration.split(' day')[0])
                return(timedelta(days = days))
            except:
                try:
                    #type '2 to 3 days'
                    days_1 = int(duration.split(' days')[0].split(' to ')[0])
                    days_2 = int(duration.split(' days')[0].split(' to ')[1]) 
                    return(timedelta(days = days_2))  
                except:
                    try:
                        #type of '5 minutes'
                        minutes = int(duration.split(' minute')[0])
                        return(timedelta(minutes = minutes)) 
                    except:
                        try:
                            #type of '4 to 5 minutes'
                            minutes_1 = int(duration.split(' minutes')[0].split(' to ')[0])
                            minutes_2 = int(duration.split(' minutes')[0].split(' to ')[1])
                            return(timedelta(minutes = minutes_2))
                        except:
                            print("FUCKED")
                            print(duration)
                    
def check_duration_type(time_obj, dur_type):
    switcher = {'one_hour' : 1 if(time_obj <= one_hour) else 0,  #ternary,
                'four_hour' : 1 if(time_obj <= four_hour) else 0,
                'one_day' :1 if(time_obj <= one_day) else 0, 
                'three_days': 1 if(time_obj <= three_days) else 0,
                'three_day_plus':1 if(time_obj > three_days) else 0}
#     print(f"######### DURATION TYPE: {dur_type}#########      #####   dur_type: {switcher.get(dur_type)}    ####")    
    return(switcher.get(dur_type))
```

```python
# import cat_subcat_dicts
# subcat = cat_subcat_dicts.cat_subcat_dicts()

def segregate_title_to_private_or_shared(title):
#     import re
    title = title.lower() #term we want to search for
    term = "Private".lower()
#     print(re.split('[:,& ()]', title))
    if term in re.split('[:,& ()]', title): #see if one of the words in the sentence is the word we want
        return "Private"
    else:
        return "Shared"
    
def remove_titles_with_airport(title):
    #     import re
    drop = 0
    title = title.lower() #term we want to search for
    term = "airport"
#     print(re.split('[:,& ()]', title))
    if term in re.split('[:,& ()]', title): #see if one of the words in the sentence is the word we want
        drop = 1
        
    return drop
```
